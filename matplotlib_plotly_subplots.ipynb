# Два графика для matplotlib и plotly

### Данные для всех графиков

```python
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2*np.pi*t)
s2 = np.sin(4*np.pi*t)
```

## matplotlib

```python
import matplotlib.pyplot as plt

plt.subplot(211)
plt.plot(t, s1)
plt.title(r'$\sin(2\pi t)$')  # LaTeX formatted title
plt.xlabel(r'$t$')  # LaTeX formatted label for X-axis

plt.subplot(212)
plt.plot(t, s2, 'g')
plt.title(r'$\sin(4\pi t)$')  # LaTeX formatted title
plt.xlabel(r'$t$')  # LaTeX formatted label for X-axis

plt.suptitle(r'$\text{Два графика синусоид}$', fontsize=20)  # LaTeX formatted super title
plt.subplots_adjust(top=0.85, hspace=0.5)  # Increase the space above the subplots and between the plots
plt.show()
```

## plotly

### plotly.graph_objs as go 

```python
from plotly.subplots import make_subplots
import plotly.graph_objs as go

fig = make_subplots(rows=2, cols=1)

fig.add_trace(
    go.Scatter(x=t, y=s1, mode='lines', name='sin(2πt)'),
    row=1, col=1
)
fig.add_trace(go.Scatter(x=t, y=s2, mode='lines', name='sin(4πt)', line=dict(color='green')), row=2, col=1)

# Add overall title
fig.update_layout(title_text='Два графика синусоид')

fig.show()
```

### plotly.express as px

```python
import plotly.express as px
from plotly.subplots import make_subplots

fig1 = px.line(x=t, y=s1, title="sin(2πt)", labels={'x': 't', 'y': 'sin(2πt)'})
fig1.update_traces(name="sin(2πt)")
fig2 = px.line(x=t, y=s2, labels={'x': 't', 'y': 'sin(4πt)'}, color_discrete_sequence=['green'])
fig2.update_traces(name="sin(4πt)")

fig = make_subplots(rows=2, cols=1)

fig.add_trace(fig1['data'][0], row=1, col=1)
fig.add_trace(fig2['data'][0], row=2, col=1)

fig.update_layout(title_text='Два графика синусоид')
fig.update_traces(showlegend=True)

fig.show()
```
